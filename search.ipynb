```python
import pickle
```

```python
with open('dictionary', 'rb') as f:
    d = pickle.load(f)
```

```python
d
```

```python
search = input("Please enter a keyword ")
if len(search.split()) >=3:
    print("Please search with one or two keywords only")
elif search.lower() in d:
    print("The keyword \'{}\' is present in \n{}.".format(search, d[search.lower()]))
else:
    print("The keyword \'{}\' is not present.".format(search))
```
